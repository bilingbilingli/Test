# 读取数据集：百度apollo红绿灯数据集：demo_data

```python
import sys
sys.path.append('/usr/local/lib/python2.7/site-packages')

import cv2
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
import matplotlib.patches as mpatches
from skimage import feature, measure,segmentation


train_path = "../demo_data/trainsets"     #路径：home/udacity/traffic_light/test/下开始，而不是从home开始
test_path = "../demo_data/testsets"

def load_dataset(imagePath): 
    for file in os.listdir(imagePath):
        fileName = os.path.join(imagePath, file)
        if os.path.isfile(fileName):
            with open(fileName, "r") as file:
                data = file.readlines()
                num = 0
                if len(data[0].rstrip().split(" ")) == 1:
                    img_list = [[0 for col in range(1)] for row in range(len(data))]
                    for row_data in data:
                        imgDir = row_data.rstrip()                        
                        imgPathList = os.path.join(imagePath, imgDir)
                        img = mpimg.imread(imgPathList)             
                        img_list[num] = img   #图像存储在img_label_list
                        num = num +1
                        file.close()
                    return img_list
                
                elif len(data[0].rstrip().split(" ")) == 2:
                    img_label_list = [[0 for col in range(2)] for row in range(len(data))]
                    for row_data in data:
                        row_data = row_data.rstrip()
                        imgDir, labDir = row_data.split(" ")
                        #print len(row_data.split(" "))
                        #print labDir
                        imgPathList = os.path.join(imagePath, imgDir)
                        img = mpimg.imread(imgPathList)
                        labPathList = os.path.join(imagePath, labDir) 
                       # file = open(labPathList);
                        text = np.loadtxt(labPathList)                 
                        img_label_list[num] = [img, text]   #图像与标签存储在img_label_list
                        num = num +1
                        file.close()
                    return img_label_list
                else:
                    print '不适用于此数据集'
    
imageList = load_dataset(train_path)
test_imageList = load_dataset(test_path)
#print imageList[0]
#imageList = load_dataset(test_path)
#print imageList[0]
```

# 提取ROI ,HOG特征向量

```python
def ROI_extra(img_label_list):
    img_ROI_list = []
    labels = []
    for i in range(len(img_label_list)):
        label_lines = int(img_label_list[i][1].size) / 5
        if label_lines ==1.0:
            label = img_label_list[i][1][0]
            x = int(img_label_list[i][1][2])
            y = int(img_label_list[i][1][1])
            maxx = int(img_label_list[i][1][4])
            maxy = int(img_label_list[i][1][3])  
            
            img_ROI = img_label_list[i][0][x:maxx, y:maxy]            
            img_ROI_list.append(img_ROI)
            labels.append(label)
        else:
            for j in range(label_lines):
                label = img_label_list[i][1][j][0]
                x = int(img_label_list[i][1][j][2])
                y = int(img_label_list[i][1][j][1])
                maxx = int(img_label_list[i][1][j][4])
                maxy = int(img_label_list[i][1][j][3])
            
                img_ROI = img_label_list[i][0][x:maxx, y:maxy]            
                img_ROI_list.append(img_ROI)
                labels.append(label)
    return img_ROI_list, np.array(labels)
    
def hog_feature(img_ROI_list):
    features = []
    for i in range(len(img_ROI_list)):
        img = cv2.cvtColor(img_ROI_list[i], cv2.COLOR_BGR2GRAY)
        img = cv2.resize(img, (64,128))
        bins = 9
        cell_size = (8, 8)
        cpb = (2, 2)
        norm = "L2"
        fts = feature.hog(img, orientations=bins, pixels_per_cell=cell_size, 
                            cells_per_block=cpb, block_norm=norm, transform_sqrt=True)
        features.append(fts)
    return features

img_ROI_list, labels = ROI_extra(imageList)
features = hog_feature(img_ROI_list)
print len(img_ROI_list)
#plt.imshow(img_ROI_list[20])
```

# SVM 分类及测试

```python
from sklearn.svm import SVC
from sklearn.externals import joblib
from sklearn.metrics import roc_curve, auc

def svm_train(features, labels, save_path="./svm_model.pkl"):

    clf = SVC(C=10, tol=1e-3, probability = True)
    clf.fit(features, labels)
    joblib.dump(clf, save_path)
    print "train finished."

def svm_test(svm_model, features, labels):

    clf = joblib.load(svm_model)
    accuracy = clf.score(features, labels)
    error_num = len(img_ROI_list)*(1 - accuracy)
    print 'dectection accuracy:', accuracy
    print 'dectection error:', error_num
    return accuracy

svm_train(features, labels, save_path="./svm_model.pkl")
svm_test("./svm_model.pkl", features, labels)
```

# 测试集提取ROI 

```python
def regionFilter(img):
    candidateRegions = []  #候选区域
    
    cleared = img.copy()  #复制
    segmentation.clear_border(cleared)  #清除与边界相连的目标物
    label_image =measure.label(cleared)  #连通区域标记
    borders = np.logical_xor(img, cleared) #异或
    label_image[borders] = -1
    region = measure.regionprops(label_image)
    if len(region) == 0:
        print '目标不存在'
    else:
        num = label_image.max()  
        del_array = np.array([0] * (num + 1))#生成一个与连通域个数相同的空数组来记录需要删除的区域（从0开始，所以个数要加1）
        save_index = 0
        for k in range(num):#TODO：这里如果遇到全黑的图像的话会报错
            #print '凸包面积', region[k].convex_area
            #忽略过小/过大区域
            if region[k].area <= 100 or region[k].area >= 3000:
                continue
                
            minr, minc, maxr, maxc = region[k].bbox
            #判断连通区域是否连接图像边界，若连接，则清除
            if minr == 0 or minc == 0 or minr+maxr == img.shape[0] or minc+maxc == img.shape[1]:
                continue
            
            w = maxc - minc
            h = maxr - minr
            Rwh =  float(w) / h
            Sat =  float(region[k].area) / region[k].convex_area
            #print 'Rwh, Sat', Rwh, Sat
            
            if Rwh >= 0.8 and Rwh <= 1.5  and Sat > 0.6:   
                #print '区域面积', region[k].area 
                
                candidateRegions.append(region[k])
                #绘制外接矩形
               # rect = mpatches.Rectangle((minc, minr), w,h,fill=False, edgecolor='yellow', linewidth=1)
                #ax0.add_patch(rect)                
                save_index = k + 1 #需要保留的索引
            del_array[save_index] = 1

        del_mask = del_array[label_image]
        out = img * del_mask
    return candidateRegions

def colorRecog(img, candidateRegions):
    colorRecogImg = img.copy() 
    #colorType = []
    regionList = [[0 for col in range(2)] for row in range(len(candidateRegions))]
    
    for i in range(len(candidateRegions)):
        numG, numR = 0, 0
        for j in range(candidateRegions[i].area): 
            x = candidateRegions[i].coords[j][0]
            y = candidateRegions[i].coords[j][1]
            if colorRecogImg[x,y] >=20 and colorRecogImg[x,y] <= 80:
                numG = numG +1
            elif colorRecogImg[x,y] >=160 and colorRecogImg[x,y] <= 220:
                numR = numR +1
            else:
                continue

        if float(numG) / candidateRegions[i].area >=0.8:
             regionList[i][1] = 'green'
        elif float(numR) / candidateRegions[i].area >=0.8:
             regionList[i][1] = 'red'
        else:
             regionList[i][1] = 'unclear'
        regionList[i][0] = candidateRegions[i]
               
    return  regionList

#向下扩展 
def downExtendRegions(x, y, w, h):     
    dn_x = int(x - 0.3*w)
    dn_y = int(y - 0.3*h)
    dn_w =int(w + 0.6*w)
    dn_h = int(h + 3.2*h)
    
    return dn_x, dn_y, dn_w, dn_h

#向上扩展
def upExtendRegions(x, y, w, h):      
    up_x = int(x - 0.3*w)
    up_y = int(y - 3*h)
    up_w =int(w + 0.6*w)
    up_h = int(h + 3.2*h)
    
    return up_x, up_y, up_w, up_h
        
  #输入：原始图像、连通区域列表
#输出：进一步筛选后的扩展区域
def  ExtendRegions(img, regionList):   
    ExRegionsFixed = []   #提取扩展后区域
    rects = []         
    imgExRegions = img.copy()
    #ROIList = [[0 for col in range(2)] for row in range(len(regionList))]
    for i in range(len(regionList)):
        
        y, x, maxy, maxx = regionList[i][0].bbox
        w = maxx - x
        h = maxy - y      
        #向上扩展[绿灯]
        if regionList[i][1] == 'green' :
            up_x, up_y, up_w, up_h = upExtendRegions(x, y, w, h)
            ExRegionsFixed.append(imgExRegions[up_y:(up_y + up_h), up_x:(up_x + up_w)])
            rects.append([up_x, up_y, up_x + up_w, up_y + up_h])
            #绘制外接矩形
           # rect = mpatches.Rectangle((up_x, up_y), up_w,up_h,fill=False, edgecolor='red', linewidth=1)
            #ax0.add_patch(rect) 
        #向下扩展[红灯]
        elif regionList[i][1] == 'red' :
            dn_x, dn_y, dn_w, dn_h = downExtendRegions(x, y, w, h)
            ExRegionsFixed.append(imgExRegions[dn_y:(dn_y + dn_h), dn_x:(dn_x + dn_w)])
            rects.append([dn_x, dn_y, dn_x + dn_w, dn_y + dn_h])
             #绘制外接矩形
            #rect = mpatches.Rectangle((dn_x, dn_y), dn_w,dn_h,fill=False, edgecolor='red', linewidth=1)
            #ax0.add_patch(rect) 
        elif regionList[i][1] == 'unclear' :
        #未扩展
            print '未辨别出颜色'
            ExRegionsFixed.append(imgExRegions[y:maxy, x:maxx])     
        #ROIList[i] = [ExRegionsFixed, regionList[i][1]]      
    #return ROIList
    return ExRegionsFixed, rects
```

```python
#输入：单帧图像RGB
def testROI(img):
    #转为YCbCr
    imgYcc = cv2.cvtColor(img, cv2.COLOR_BGR2YCR_CB)
    #对Cb通道 二值化
    lower = np.array([80])
    upper = np.array([160])
    imgYcc_thresh = cv2.inRange(imgYcc[:, :, 2], lower, upper)
    imgYccThresh = 255-imgYcc_thresh   
    #plt.imshow(imgYccThresh, cmap = 'gray')
    
    candRegs = regionFilter(imgYccThresh)
    regions = colorRecog(imgYcc[:,:,2], candRegs)  #[候选区域，对应颜色]
    ROI,rects = ExtendRegions(img, regions)   
    return ROI, regions, rects

#ROI, regions, rect = testROI(test_imageList[0])
#print '检测到的交通信号灯数量：', len(rect)
#plt.imshow(ROI[0])
```
